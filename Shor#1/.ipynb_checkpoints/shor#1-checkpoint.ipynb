Erster Versuch der Implementierung des Shor Algorithmus 
Die Implementierung faktoriert die Zahl N=15. Aufgrund einer besonderen eigenschaft von N kann die modular exponential Funktion in diesem Fall vereinfacht werden wodurch diese Berechnung nur 4 Qubits benötigt.
Diese Variante kann **nicht** für allgemeine N genutzt werden.

![image](resources/qpe.svg)

Die Grafik oben(Quelle: https://en.wikipedia.org/wiki/Quantum_phase_estimation_algorithm 01.06.2023) zeigt eine Grafik des so genannten "Quantum Phase Estimation" Algorithmus. Zweck des Algorithmus ist die Bestimmung der Phase des Eigenvektors $\vert\psi\rangle$ durch eine Unitäre Operation (Hier U).

![image](resources/Shors_algorithm.svg)
Zeigt die Basisversion des Shor Algorithmus. Der Aufbau ähnelt dem der QPE, mit Ausnahme das 2n-1 anstatt n-1 U Operationen notwenig sind. Darum wird Shor auch als "Quantum Phase Estimation in disguise" bezeichnet. 
Aufgrund der Vereinfachung des Shor beim N=15 sind in diesem Falle n-1 anstatt 2n-1 U transformationen notwenig.

```python
import qiskit 
import numpy as np
from qiskit.visualization import plot_histogram

my_gate_names = {
    "displaytext": {
        # gate name : gate text 
        "x": "Pauli X",
        "y": "Pauli Y",
        "z": "Pauli Z",
        "h": "Hadamard",
        "reset": "Reset",
        "id": "Identity",
    },
}
def initialize_qubits(given_circuit, n, m):
    for i in range(n):
      given_circuit.h(i)
    #for i in range(n,n+m):
    given_circuit.x(m+n - 1)

#Nutzt ein Phasengate(https://qiskit.org/documentation/stubs/qiskit.circuit.library.PhaseGate.html)
#Fügt nur eine Phase hinzu und verändert die Amplitude nicht
def unitary_operator(theta):
    return (qiskit.circuit.library.standard_gates.PhaseGate(theta,str(theta))).control(1)

def apply_unitary_operator(quantum_circuit, theta, control, target, exponent):
    #Rechne theta * exponent da schneller als exponent-Fache anwendung von U
    quantum_circuit.append(unitary_operator(theta/(2**exponent)),[control, target])

#Die standart QFT sollte durch die approximate quantum fourier transform ersetzt werden. 
def myQFT(quantum_circuit, measurement_qubits, inverse = False, swaps = False):
    if not inverse:
        for target in measurement_qubits:
            quantum_circuit.barrier()
            exponent = 2
            quantum_circuit.h(target)
            for control in measurement_qubits[target + 1:]:
                apply_unitary_operator(quantum_circuit, 2 * np.pi, control, target, exponent)
                exponent+=1
        quantum_circuit.barrier()
        if swaps:
            for x in range(0, len(measurement_qubits)//2):
                quantum_circuit.swap(measurement_qubits[x], measurement_qubits[len(measurement_qubits) - 1 - x])
                x+=1
        quantum_circuit.barrier()
    else:
        if swaps:
            quantum_circuit.barrier()
            for x in range(0, len(measurement_qubits)//2):
                quantum_circuit.swap(measurement_qubits[x], measurement_qubits[len(measurement_qubits) - 1 - x])
                x+=1
        for target in reversed(measurement_qubits):
            quantum_circuit.barrier()
            exponent = len(measurement_qubits[target + 1:]) + 1
            for control in reversed(measurement_qubits[target + 1:]):
                apply_unitary_operator(quantum_circuit, -2 * np.pi, control, target, exponent)
                exponent -= 1
            quantum_circuit.h(target)
        quantum_circuit.barrier()
```

```python
def a_x_mod15(a, x):
    if a not in [2,7,8,11,13]:
        raise ValueError("'a' must be 2,7,8,11 or 13")
    U = qiskit.QuantumCircuit(4)        
    for iteration in range(x):
        if a in [2,13]:
            U.swap(0,1)
            U.swap(1,2)
            U.swap(2,3)
        if a in [7,8]:
            U.swap(2,3)
            U.swap(1,2)
            U.swap(0,1)
        if a == 11:
            U.swap(1,3)
            U.swap(0,2)
        if a in [7,11,13]:
            for q in range(4):
                U.x(q)
    U = U.to_gate()
    U.name = "%i^%i mod 15" % (a, x)
    c_U = U.control()
    return c_U

def modular_exponentiation(quantum_circuit, measurement_qubits, n, a):
    for x in range(measurement_qubits):
        exponent = 2**x
        quantum_circuit.append(a_x_mod15(a, exponent), 
                     [x] + list(range(measurement_qubits, measurement_qubits+n)))

```

```python
measurement_qubits = 4
target_qubits = 4
a = 7
qc = qiskit.QuantumCircuit(measurement_qubits + target_qubits, measurement_qubits)
initialize_qubits(qc, measurement_qubits, target_qubits)
qc.barrier()
modular_exponentiation(qc, measurement_qubits, target_qubits, a)
myQFT(qc,[0,1,2,3],swaps = False,inverse= True)
qc.barrier()
qc.measure(range(4),range(4))
qc.draw('mpl', style=my_gate_names, scale = 0.75)
```

```python
simulator = qiskit.Aer.get_backend('aer_simulator')
counts = qiskit.execute(qc, backend=simulator, shots=1000).result().get_counts(qc)
plot_histogram(counts)
```

```python
from math import gcd

for measured_value in counts:
    measured_value_decimal = int(measured_value[::-1], 2)
    print(f"Gemessen {measured_value_decimal}")
    
    if measured_value_decimal % 2 != 0:
        print("Messwert ist keine gerade Zahl")
        continue
    x = int((a ** (measured_value_decimal/2)) % 15)
    if (x + 1) % 15 == 0:
        print("Fehler da x + 1 = 0 (mod N) bei x = a^(r/2) (mod 15)")
        continue
    guesses = gcd(x + 1, 15), gcd(x - 1, 15)
    print(guesses)
```
